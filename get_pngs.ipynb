```python
from PIL import Image
import numpy as np
import os
```

```python
regions = sorted(os.listdir('scenes_allbands/'))
images = []
for (i, region) in enumerate(regions):
    image = np.load('scenes_allbands/' + region)
    image = image[:, :, [3, 5, 6]]
    images.append(image)
```

```python
images = [Image.fromarray(image) for image in images]
```

```python
images[0]
```

```python
for i, image in enumerate(images):
    image.save(f'scenes_467/467_x{i + 1 :02d}.png')
```
